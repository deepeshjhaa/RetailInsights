```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data=pd.read_excel("sales_data.xlsx")
df=pd.DataFrame(data)

df["Month"] = df["Date"].dt.strftime("%B")  
ms=df.groupby("Month")["Total"].sum().reset_index()
plt.figure(facecolor='lightblue')
sns.set_style("darkgrid")
# print(ms)
bars=ms.plot(kind="bar", color="red", edgecolor="black")
plt.gca().set_facecolor("mintcream") 
for bar in bars.patches:
    bars.text(bar.get_x() + bar.get_width()/2,
              bar.get_height(),
              int(bar.get_height()),
              ha='center', va='bottom')

plt.title("Monthly Sales Trend", color='darkgreen', fontsize=15, fontweight='bold', fontname='Times New Roman')
plt.xlabel("Months",color='blue', fontsize=12, fontweight='bold', fontname='Arial')
plt.ylabel("Total Sales",color='blue', fontsize=12, fontweight='bold', fontname='Arial')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.grid(True)
# plt.tight_layout()
# plt.grid(True)
plt.savefig("monthly_sales_chart.png") 
plt.show()
print(". Total Sales:", df["Total"].sum())

ds=df.groupby("Product")["Quantity"].sum().reset_index()
plt.figure(facecolor='lightgray')
sns.set_style("darkgrid")
# print(ms)
lines=ds.plot( color="red", marker="*",markerfacecolor="black", markeredgecolor="black")
plt.gca().set_facecolor("steelblue") 

plt.title("Top Selling Product", color='blue', fontsize=15, fontweight='bold', fontname='Times New Roman')
plt.xlabel("Product",color='green', fontsize=12, fontweight='bold', fontname='Arial')
plt.ylabel("Quantity",color='green', fontsize=12, fontweight='bold', fontname='Arial')
plt.xticks(rotation=0)
plt.yticks(rotation=0)

# plt.tight_layout()
plt.grid(True)
plt.savefig("top_selling_products_chart.png")
plt.show()



ms.columns = ["Month", "Total Sales"]
ds.columns = ["Product", "Quantity Sold"]
pd.concat([ms, ds], axis=1).to_csv("combined_sales_report.csv", index=False)
```
